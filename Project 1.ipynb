```python
import pandas as pd
```

```python
df = pd.read_csv("data_science_salaries.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], bins=30, kde=True)
plt.title('Distribution of salaries in US dollars')
plt.xlabel('Salary in dollars')
plt.ylabel('Number of developers')
plt.show()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salaries according to experience level')
plt.xlabel('experience_level')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='work_year', y='salary_in_usd', data=df)
plt.title('The impact of remote work on salaries')
plt.xlabel('work_year')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=df)
plt.title('Salaries by company size')
plt.xlabel('company_size')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='company_location', y='salary_in_usd', data=df)
plt.title('Salaries according to the employee')
plt.xlabel('company_location')
plt.ylabel('salary_in_usd')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=df)
plt.title('Salaries by type of employment')
plt.xlabel('employment_type')
plt.ylabel('salary_in_usd')
plt.show()
```
